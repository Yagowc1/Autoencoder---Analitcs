# Importações

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

# Pré-processamento

```python
# Lê os dados da planilha e retira somente as colunas numéricas
df = pd.read_csv('computer_prices_all.csv')
df = df[['release_year', 'cpu_tier', 'cpu_cores', 'cpu_threads', 'cpu_base_ghz', 'cpu_boost_ghz', 'gpu_tier', 'vram_gb', 'ram_gb', 'storage_gb', 'storage_drive_count', 'display_size_in', 'refresh_hz', 'battery_wh', 'charger_watts', 'psu_watts', 'bluetooth', 'weight_kg', 'warranty_months', 'price']]

# Centraliza e escalona os dados, para que possam ser usados no PCA
scaled_data = preprocessing.scale(df.T)
# Após centralizar os dados, a média dos valores para cada coluna será próxima 0
# Após escalonar os dados, o desvio padrão para os valores de cada coluna será próximo 1
```

# Implementação do PCA

```python
# Implementa o PCA
pca = PCA()
pca.fit(scaled_data)
# Gera coordenadas para um gráfico PCA baseado nos dados obtidos
pca_data = pca.transform(scaled_data)
```

# Gráfico de barras

```python
# Calcula a porcentagem de variação que cada Componente Principal representa na base de dados
per_var = np.round(pca.explained_variance_ratio_*100, decimals=1)
# Gera os rótulos de cada PCA no gráfico (ex.: PC1, PC2, PC3 e etc.)
labels = ['PC' + str(x) for x in range(1, len(per_var)+1)]

# Gera e exibe um gráfico de barras que representa a porcentagem de variação de cada Componente Principal
plt.bar(x=range(1, len(per_var)+1), height=per_var, tick_label=labels)
plt.ylabel('Porcentagem da Variação Explicada')
plt.xlabel('Componente Principal')
plt.title('Scree Plot')
plt.show()
```

# Gráfico de dispersão

```python
# Coloca os dados gerados pela PCA e colocamos em uma matriz
pca_df = pd.DataFrame(pca_data, columns=labels)

# Gera e exibe um gráfico de dispersão que usa somente 2 dos 20 Componentes Principais (já que as duas unidas representam mais de 95% da variância dos 
# dados), permitindo que o gráfico seja 2D
plt.scatter(pca_df.PC1, pca_df.PC2)
plt.title("Gráfico PCA")
plt.xlabel('PC1 - {0}%'.format(per_var[0]))
plt.ylabel('PC2 - {0}%'.format(per_var[1]))
plt.show()
```

# Quais instâncias tiveram mais impacto em separar os dados do gráfico?

```python
# Cria um objeto Series do Pandas com os Loading Scores do Componente Principal 1 (PC1)
loading_scores = pd.Series(pca.components_[0])
# Ordena os Loading Scores baseado em sua magnitude (valor absoluto)
loading_scores_ordenados = loading_scores.abs().sort_values(ascending=False)

# Pega as 50 instancias com mais impacto no gráfico e as apresenta
top_50_instancias = loading_scores_ordenados[0:50].index.values
print(loading_scores[top_50_instancias])
# Obs.: Devido ao número massivo de dados, os valores ainda estarão bem próximos
```
